```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("farhanmd29/50-startups")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/50-startups/50_Startups.csv')
```

```python
print(df)
```

```python
X = df.iloc[:, 1:-1].values
Y = df.iloc[:, -1].values

print(X)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [2])],
    remainder='passthrough'
)
X = ct.fit_transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lr =  LinearRegression()

lr.fit(x_train, y_train)
```

```python
y_prid = lr.predict(x_test)
print(y_test)
```

```python
import numpy as np
np.set_printoptions(precision=2)
np.concatenate((y_test.reshape(len(y_test), 1), y_prid.reshape(len(y_prid), 1)), axis=1)
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("satyajeetrai/medical-cost")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/medical-cost/dataset_.csv')
df.columns = df.columns.str.replace(" ", "_")
print(df)
```

```python
print(df['region'].unique())
print(df['sex'].unique())
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1].values

print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['smoker'] = le.fit_transform(X['smoker'])
print(X['smoker'])
```

```python
print(X)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [1,5])],
    remainder='passthrough'
)
X = ct.fit_transform(X)
```

```python
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size = 0.5, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
import numpy as np
np.set_printoptions(precision=2)
np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_pred), 1)), axis=1)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
